1. **Story** The story of the chart is McDonald's dominance of the fast food industry although Starbucks feels like a "one of these things is not like the other". If we think in terms of "Expressing Quantitative Relationships with Charts" we're definitely looking at *ranking* here.
2. **Problems** The main principle that is being violated is that the "bars" are not fixed and constant widths. We perceive the logos as *areas* and so although McDonald's $41bn in sales is only 4x that of Burger King, it looks at least 8x to 12x bigger. It's also not entirely clear what the inclusion of Afghanistan's GDP means. The sales of a fast food restaurant and GDP are really different things and it's also not clear if the target reader has a sense of the scale of Afghanistan's GDP! If you're going to include a comparison, it has to be something that the target reader can really compare *to*. It's tempting to always say "red and green color blindness" but that only really matters if red and green are the only or predominant colors and especially not if it's only red or only green.
3. **Alternatives** We could do this as a table but the goal here is to compare values, not to read them. As a story about ranking, we could just do a real bar chart. A bar chart will fix the problem with distortions using height *and width* that come from using images as bars.

    It then becomes interesting to think about whether we should do ascending or descending (or both...it's ok to chart the same data different ways to tell slightly different stories). We could also change the story a bit and talk about *deviation*. We could also look up the total fast food market (and perhaps update the numbers) and talk about *part/whole* relationships and market share. Same data, at least three different stories.

Let's do a bar chart, in the same order to start. We need to make sure everything is labeled correctly.

```python
sales = [4.1, 4.3, 8.0, 8.2, 9.4, 11.3, 41.0]
width = 1/1.5
xs = list(range(len( sales)))
x_labels = ["Starbucks", "Taco Bell", "Pizza Hut", "KFC", "Wendy's", "Burger King", "McDonald's"]

figure, axes = plt.subplots()

axes.set_title("Fast Food Sales")
axes.bar(xs, sales, width, color="dimgray", align="center")
axes.set_xticks(xs)
axes.set_xticklabels(x_labels)
axes.xaxis.grid(b=False)
axes.set_xlabel("Chain")
axes.set_ylabel("US dollars (billions)")

plt.show()
plt.close()
```

So that's a good start except that our labels overlap.

**Note** - You should just make the fix I'm about to make "in place". I'm describing how I fixed it to you so you know what to do and how to figure out things for yourself. 

Here's the search:

https://www.google.com/search?q=matplotlib+x+labels+rotate&oq=matplotlib+x+labels+rotate

and in the results, I found a Stackoverflow article on rotating the x labels in matplotlib. It uses the command version of Matplotlib but I was able to figure out that the `rotation=45` part goes in `set_xticklabels`. You might have also needed to look at the documentation for Matplotlib.

```python
sales = [4.1, 4.3, 8.0, 8.2, 9.4, 11.3, 41.0]
width = 1/1.5
xs = list(range(len( sales)))
x_labels = ["Starbucks", "Taco Bell", "Pizza Hut", "KFC", "Wendy's", "Burger King", "McDonald's"]

figure, axes = plt.subplots()

axes.set_title("Fast Food Sales")
axes.bar(xs, sales, width, color="dimgray", align="center")
axes.set_xticks(xs)
axes.set_xticklabels(x_labels, rotation=45)
axes.xaxis.grid(b=False)
axes.set_xlabel("Chain")
axes.set_ylabel("US dollars (billions)")

plt.show()
plt.close()
```

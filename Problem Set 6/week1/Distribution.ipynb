```python
# Normal Distribution
# Distribution of baby carrot

data = np.array([8, 8.2, 6.2, 6.4, 8.5, 7, 6.9, 6.1, 7.5, 6.2, 6.8, 
                 6.5, 7.3, 6.3, 7.2, 7.9, 9.3, 9, 9.7, 9.7, 10.3, 8, 
                 7.2, 7.4, 6.8, 4.7, 5.4, 7.6, 7.5, 6.8, 6.5, 4.2])

mean = np.mean(data)
std = np.std(data)
x = np.arange(2, 12, 0.1)

figure = plt.figure(figsize=(9, 4))

axes = figure.add_subplot(1, 1, 1)

axes.hist(data, density=True, bins="sqrt", color="steelblue")
axes.set_xlabel("Sample Mean Carrot Mass (g)")
axes.set_ylabel("Density")
axes.set_title("Distribution of Baby Carrot Weights")
axes.plot(x, norm.pdf(x, loc=mean, scale=std), 'r-', lw=1)
axes.plot()

plt.show()
plt.close()
```
